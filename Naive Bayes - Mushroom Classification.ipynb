## Naives Bayes Classification -  Mushroom Classification using Naives Bayes Model

### Loading the Dataset

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('mushroom.csv')
df.head(n=10)
```

### Encoding the categorical data into numerical data

```python
le = LabelEncoder()
## applies transformation on each column
# it basically applies encoding to each column from categorical data, by default in pandas apply(), its axis is 0, its means it applies on each columns
#fit_transform applies encoding or label on array
ds = df.apply(le.fit_transform)
# ds has integer label instead of categorical label
type(ds)
# it is again pandas dataframe
ds.head(n=10)
```

```python
# now we are done to convert it into numpy array,
data = ds.values # ---> we are not giving brackets because, it is property not function
print(data.shape)
print(type(data))
print(data[:5,:])
#now data is ready, we are ready for further computation
# if you observe carefully, you will see, the zeroth column of dataframe is y feature and rest are x features now split them
data_x = data[:,1:]
data_y = data[:,0]
```

### Break the data into train and test

```python
x_train,x_test,y_train,y_test = train_test_split(data_x,data_y,test_size = 0.2)
```

### Building Our Classifier

```python
# here is shortcut to count labels
a  =np.array([1,1,1,0,0,0])
np.sum(a==1)
# its returns a value which mention total number of labels habing value one
```

```python
def prior_prob(y_train,label):
    total_examples = y_train.shape[0]
    class_examples = np.sum(y_train==label)
    
    return (class_examples)/float(total_examples)
```

```python
# in condtional probabilty---> first we short out the class which we want, no. of outcomes from all outcomes having class X
# then after this filtered outcome we find in how many outcomes having label Y 
def cond_prob(x_train,y_train,feature_col,feature_val,label):
    x_filtered = x_train[y_train==label]
    numerator = np.sum(x_filtered[:,feature_col]==feature_val)
    denominator = np.sum(y_train==label)
    
    return numerator/float(denominator)
```

```python
### computing posterior probability
```

```python
def predict(x_train,y_train,xtest):
    classes = np.unique(y_train)
    n_features = x_train.shape[1]
    
    post_prob = [] #list of prob for all the classes and given a single testing point
    for label in classes:
        likelihood = 1.0
        for f in range(n_features):
            cond = cond_prob(x_train,y_train,f,xtest[f],label)
            likelihood*=cond
            
        prior = prior_prob(y_train,label)
        
        post = likelihood*prior #--> neglecting marginal probability
        post_prob.append(post)
    pred = np.argmax(post_prob)
    return pred
            
```

```python
output = predict(x_train,y_train,x_test[1])
print(output)
print(y_test[1])
```

```python
def score(x_train, y_train,x_test,y_test):
    pred = []
    for i in range(x_test.shape[0]):
        pred_label = predict(x_train,y_train,x_test[i])
        pred.append(pred_label)
        
    pred = np.array(pred)
    acc = np.sum(pred==y_test)/y_test.shape[0]
    return acc
```

```python
print(score(x_train,y_train,x_test,y_test))
```
